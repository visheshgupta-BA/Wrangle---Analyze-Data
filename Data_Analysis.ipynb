## Annaul Growth for 2018

```python
fig = px.bar(df1.query('Year==2018'),
y = 'value',
x= 'Country Name',
color="Country Name", 
template="plotly_dark")

fig.update_layout(title_text="Data: the World Bank" + '<br>' + "www.worlddev.xyz", title_x=0.5)
fig.update_yaxes(title = '2018 Adjusted Net National Income' + '<br>' + '(Annual % Growth)')
fig.update_layout(showlegend=True)


fig.show()
```

## As depicted in this chart, we can clearly see that India and Russia had exellent growth rate at about 6% followed by China, USA and UK. Brazil had the lowest growth rate.

## Annaul Growth for 2020

```python
fig = px.bar(df1.query('Year==2020'),
x= 'Country Name',
y='value',
color = 'Country Name',
template= 'plotly_dark')

fig.update_layout(title_text="Data: the World Bank" + '<br>' + "www.worlddev.xyz", title_x=0.5)
fig.update_yaxes(title = '2020 Adjusted Net National Income' + '<br>' + '(Annual % Growth)')
fig.update_layout(showlegend=True)

fig.show()
```

## In this chart we can see that covid happened in 2020 and most countries enforced lockdowns, due to which the gdp growth is negative for all the countries. India was hit the worst with more that -7% growth and China was least affected.

## Annaul Growth for 2020 (with Pie Chart)

## Here is same data but with Pie chart representation, Pie chart really helps when we want to see individual contributions of limited entities. As we can see here that when all the gdp growth combined, India and Russia constitutes more than half the growth amongst these countries.

## Annual Growth for 2005 - 2020

```python
df3 = df.iloc[0:4939, :]
df3 = df3[(df3['Country Name'] == 'India') | (df3['Country Name'] == 'Russian Federation') | (df3['Country Name'] == 'China') | (df3['Country Name'] == 'United States') | (df3['Country Name'] == 'United Kingdom') | (df3['Country Name'] == 'South Africa') | (df3['Country Name'] == 'Brazil')]
df3 = df3[df3['Year'].between(2005,2020)]

fig = px.line(df3,
x= 'Year',
y='value',
color = 'Country Name',
template="plotly_dark")

fig.update_layout(title_text="Data: the World Bank" + '<br>' + "www.worlddev.xyz", title_x=0.5)
fig.update_yaxes(title = '2018 Death Rate, Crude (per 1,000 people)')
fig.update_layout(showlegend=True)

fig.show()
```

## This is the same chart but with more details for recent years. Russia's gdp growth looks very volatile as a lot of fluctuations can be seen, we can also see that almost all countries went down in 2008 when the american economy collapsed, which tells how different countries economy is impacted when there is a recession in one of the major economies.
